## Observations and Insights 

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import scipy.stats as st
from scipy.stats import linregress


# Study data files
mouse_metadata_path = "Resources/Mouse_metadata.csv"
study_results_path = "Resources/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)
```

```python
study_results['Mouse ID'].value_counts()
```

```python
# Combine the data into a single dataset
mouse_study = pd.merge(mouse_metadata, study_results, on='Mouse ID', how='left')
pd.options.display.float_format = '{:,.2f}'.format
# Display the data table for preview
mouse_study.head()
```

```python
# Checking the number of mice.
mouse_study['Mouse ID'].nunique()
```

```python
mouse_study.shape
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
duplicates = [mouse_study.groupby(['Mouse ID', 'Timepoint']).count()>1][0]
duplicates[duplicates.index.get_level_values(0)=='g989']
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
mouse_study.duplicated(subset=['Mouse ID', 'Timepoint']).value_counts()

```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
mouse_study.set_index('Mouse ID', inplace=True)
mouse_study =mouse_study.drop('g989', axis=0)
```

```python
# Checking the number of mice in the clean DataFrame.
mouse_study.shape
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
regimen_summary = mouse_study.groupby('Drug Regimen')
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
tumor_mean = regimen_summary['Tumor Volume (mm3)'].mean()
tumor_median =regimen_summary['Tumor Volume (mm3)'].median()
tumor_variance = regimen_summary['Tumor Volume (mm3)'].var()
tumor_std =regimen_summary['Tumor Volume (mm3)'].std()
tumor_sem =regimen_summary['Tumor Volume (mm3)'].sem()
# Assemble the resulting series into a single summary dataframe.
summary_stats = pd.DataFrame({'Mean': tumor_mean, 'Median': tumor_median, 'Variance': tumor_variance, 'Standard Deviation': tumor_std, 'SEM': tumor_sem})
summary_stats
```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Using the aggregation method, produce the same summary statistics in a single line
regimen_summary['Tumor Volume (mm3)'].agg(['mean','median','var','std','sem'])
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin

# Start by getting the last (greatest) timepoint for each mouse
group_mice = mouse_study.groupby(['Mouse ID','Drug Regimen'])
max_tp = group_mice.max('Timepoint')
rename_max = max_tp.rename(columns={'Timepoint':'Max Timepoint'})
new_tp_df = rename_max[['Max Timepoint', 'Tumor Volume (mm3)']]

# Merge this group df with the original dataframe to get the tumor volume at the last timepoint
timepoint_merge = pd.merge(new_tp_df, mouse_study, on='Mouse ID')
```

```python
# Start by getting the last (greatest) timepoint for each mouse
# Merge this group df with the original dataframe to get the tumor volume at the last timepoint - 
#no need to merge, filtered by largest timepoint
group_mice = mouse_study.groupby(['Mouse ID','Drug Regimen'])
max_tp = group_mice.max('Timepoint')
rename_max = max_tp.rename(columns={'Timepoint':'Max Timepoint',
                                   'Tumor Volume (mm3)':'Tumor Volume at Max TP'})
new_tp_df = rename_max[['Max Timepoint', 'Tumor Volume at Max TP']]
new_tp_df.head()
```

```python
# Merge this group df with the original dataframe to get the tumor volume at the last timepoint
timepoint_merge = pd.merge(new_tp_df, mouse_study, on='Mouse ID')
timepoint_merge.head()
```

```python
# Put treatments into a list for for loop (and later for plot labels)
treatment_list= ['Capomulin', 'Ramicane', 'Infubinol', 'Ceftamin']
treatment_values = []
four_treatments = timepoint_merge[timepoint_merge['Drug Regimen'].isin(treatment_list)]
four_treatments.head()
```

```python
# Create empty list to fill with tumor vol data (for plotting)
four_max_tumor = four_treatments.groupby(['Drug Regimen', 'Mouse ID']).agg(\
                                                              max_tumor_size=('Tumor Volume (mm3)', lambda x: x.iloc[-1]))
max_tumor = four_max_tumor.stack(level=0).unstack(level=0)
max_tumor.head()
```

```python
max_tumor = four_max_tumor.stack(level=0).unstack(level=0)
max_tumor.head()
```

```python
# Put treatments into a list for for loop (and later for plot labels)
#set divider
loop_divider = 0

# Create empty list to fill with tumor vol data (for plotting)
for treatment in treatment_list:
    quartiles = max_tumor[treatment].quantile([.25,.5,.75]).round(2)
    lowerq = quartiles[0.25].round(2)
    upperq = quartiles[0.75].round(2)
    iqr = round(upperq-lowerq,2)
    lower_bound = round(lowerq - (1.5*iqr),2)
    upper_bound = round(upperq + (1.5*iqr),2)
    count_outliers = (four_treatments['Tumor Volume (mm3)'].loc[(four_treatments['Tumor Volume (mm3)'] >= upper_bound) | 
                                        (four_treatments['Tumor Volume (mm3)'] <= lower_bound)]).count()
    



    
    if loop_divider == 0:
        print(f'------------------------------------')        
    print(f'{treatment} Data:')
    print(f'IQR:{iqr}')
    print(f'Lower Quartile:{lowerq}')
    print(f'Upper Quartile:{upperq}')
    print(f'Lower Quartile Boundary:{lower_bound}')
    print(f'Upper Quartile Boundary:{upper_bound}')
    print(f'Values below {lower_bound} or above {upper_bound} could be outliers.')
```

```python
# Generate a box plot of the final tumor volume of each mouse across four regimens of interest
box_list = []
for treatment in treatment_list:
    box_list.append(list(max_tumor[treatment].dropna()))
```

```python
# Generate a box plot of the final tumor volume of each mouse across four regimens of interest
fig1, ax = plt.subplots(figsize=(10,8))
ax.set_title('Final Tumor Volume by Regimen')
ax.set_xlabel('Drug Regimen')
ax.set_ylabel('Final Tumor Volume')
ax.boxplot(box_list)
plt.xticks([1,2,3,4], treatment_list)
plt.show()
plt.savefig("Results/boxplot.png")
plt.show()
```

## Line and Scatter Plots

```python
# Generate a line plot of tumor volume vs. time point for a mouse treated with Capomulin
capomulin = mouse_study.loc[mouse_study['Drug Regimen'] == 'Capomulin']
capomulin.head(20)
```

```python
# Generate a line plot of tumor volume vs. time point for a mouse treated with Capomulin
#select mouse
capomulin_mouse = mouse_study.loc[mouse_study.index == 'x401']
#generate line plot
plt.plot(capomulin_mouse['Timepoint'], capomulin_mouse['Tumor Volume (mm3)'], marker = 'o')

#add Title, xlabel, ylabel
plt.title('Capomulin Tumor Vol vs Time Point for Mouse x401')
plt.xlabel('TimePoint')
plt.ylabel('Tumor Volume (mm3)')
plt.savefig("Results/Capomulin_tumor_v_time_line.png")
plt.show()
```

```python
# Generate a scatter plot of average tumor volume vs. mouse weight for the Capomulin regimen
capomulin_by_mouse = capomulin.groupby(['Mouse ID']).agg(\
                                                        avg_tumor_vol =('Tumor Volume (mm3)', np.mean),\
                                                        avg_mouse_wgt =('Weight (g)', np.mean)\
                                                        ).round(3)
capomulin_by_mouse.head(30)
```

```python
plt.scatter(capomulin_by_mouse['avg_mouse_wgt'],capomulin_by_mouse['avg_tumor_vol'], marker='o')
plt.title("Capomulin Regimen Weight vs Tumor Volume")
plt.xlabel('Weight')
plt.ylabel('Tumor Volume')
plt.savefig("Results/Capomulin_scatter.png")
plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and linear regression model 
# for mouse weight and average tumor volume for the Capomulin regimen
#correlation coefficient
cor_co = st.pearsonr(capomulin_by_mouse['avg_mouse_wgt'],capomulin_by_mouse['avg_tumor_vol'])
print(f'The correlation coefficient between mouse weight and average tumor volume is {round(cor_co[0],2)}.')
#linear regression model
xvalues =capomulin_by_mouse['avg_mouse_wgt']
yvalues = capomulin_by_mouse['avg_tumor_vol']
(slope, intercept, rvalue, pvalue, stderr)= linregress(xvalues, yvalues)
rvalues = xvalues * slope + intercept
linear_eq = f'y = {str(round(slope,2))}x + {str(round(intercept,2))}'
plt.scatter(xvalues, yvalues)
plt.plot(xvalues, rvalues, "r-")
plt.title('Weight vs Tumor Volume')
plt.xlabel('Weight (g)')
plt.ylabel('Tumor Volume (mm3)')
plt.annotate(linear_eq,(19,36), fontsize=14)
plt.savefig("Results/linear_regression.png")
plt.show()
```
